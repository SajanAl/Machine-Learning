```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
```

```python
df=pd.read_csv('D:/Data/titanic.csv')
```

```python
df0=df[['Age','Pclass','SibSp','Parch','Survived']]
```

```python
df0
```

```python
df0.dropna(inplace=True)
```

```python
df0.isna().sum()
```

```python
X=df0.drop(['Survived'],axis='columns')
Y=df0['Survived']
```

```python
cross_val_score(LogisticRegression(),X,Y,scoring='accuracy',cv=20).mean()
```

# Apply feature consturction

```python
X['Family_size']=X['SibSp']+X['Parch']+1
```

```python
X
```

```python
def myfunc(num):
    if num==1:
        return 0
    elif num >1 and num<=4:
        return 1
    else:
        return 2
```

```python
myfunc(4)
```

```python
X['Family_type']=X['Family_size'].apply(myfunc)
```

```python
X
```

```python
X.drop(columns=['Family_size'],inplace=True)
```
