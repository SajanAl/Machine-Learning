```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
df=pd.read_csv("D:/Data/wine_data.csv",header=None,usecols=[0,1,2])
df.columns=["Class label","Alchol","Malic Acid"]
```

```python
df.head(5)
```

```python
sns.kdeplot(df['Alchol'])
```

```python
sns.kdeplot(df['Malic Acid'])
```

```python
color={1:'red',2:'green',3:'blue'}
sns.scatterplot(data=df,x='Alchol',y='Malic Acid',hue='Class label',palette=color)
```

```python
X=df.drop(['Class label'],axis='columns')
Y=df['Class label']
```

```python
X.head(5)
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=0)
```

```python
X_test.head(5)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
model=MinMaxScaler()
```

```python
model.fit(X_train)
```

```python
X_train_scaled=model.transform(X_train)
X_test_scaled=model.transform(X_test)
```

```python
x_train_scaled=pd.DataFrame(X_train_scaled,columns=X_train.columns)
x_test_scaled=pd.DataFrame(X_test_scaled,columns=X_test.columns)
```

```python
x_train_scaled
```

```python
#Before Scalling 
np.round(X_train.describe(),1)
```

# After Scalling

```python
np.round(x_train_scaled.describe(),1)
```

# Now plot the graph

```python
fig,(ax1,ax2)=plt.subplots(ncols=2,figsize=(12,5))
ax1.set_title("Before Scalling")
ax1.scatter(data=X_train,x='Alchol',y='Malic Acid',c=Y_train)
ax2.set_title("After Scalling")
ax2.scatter(data=x_train_scaled,x='Alchol',y='Malic Acid',c=Y_train)
```

# Now plot kdeplot graph

```python
fig,(ax1,ax2)=plt.subplots(ncols=2,figsize=(12,5))
ax1.set_title("Before Scalling")
sns.kdeplot(X_train['Alchol'],ax=ax1)
sns.kdeplot(X_train['Malic Acid'],ax=ax1)
ax2.set_title("After Scalling")
sns.kdeplot(x_train_scaled['Alchol'],ax=ax2)
sns.kdeplot(x_train_scaled['Malic Acid'],ax=ax2)
```

# comparision Before and After

```python
fig,(ax1,ax2)=plt.subplots(ncols=2,figsize=(12,5))
ax1.set_title("Before Distribution Alchol")
sns.kdeplot(X_train['Alchol'],ax=ax1)
ax2.set_title("After Distribution Alchol")
sns.kdeplot(x_train_scaled['Alchol'],ax=ax2)
```

```python
fig,(ax1,ax2)=plt.subplots(ncols=2,figsize=(12,5))
ax1.set_title("Before Distribution Malic Acid")
sns.kdeplot(X_train['Malic Acid'],ax=ax1)
ax2.set_title("After Distribution Malic Acid")
sns.kdeplot(x_train_scaled['Malic Acid'],ax=ax2)
```
