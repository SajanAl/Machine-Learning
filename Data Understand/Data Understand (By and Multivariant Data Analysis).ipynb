```python
import seaborn as sns
import pandas as pd
```

```python
tips=sns.load_dataset('tips')
tips.sample(5)
```

```python
titanic=pd.read_csv('D:/Data/train.csv')
titanic.sample(5)
```

```python
flight=sns.load_dataset('flights')
flight.sample(5)
```

```python
iris=sns.load_dataset('iris')
iris.sample(5)
```

# By Data Analysis

# 1 Scatterplot(Numrical-Numrical) find releation b/w them

```python
sns.scatterplot(data=tips,x='total_bill',y='tip')
```

# Multivariant Data Analysis

```python
sns.scatterplot(data=tips,x='total_bill',y='tip',style='sex',hue='smoker')
```

```python
sns.relplot(data=tips,x='total_bill',y='tip',hue='day',style='day',col='time',kind='scatter')
```

# 2 Bar Plot(Numrical-Categorical)

```python
sns.barplot(data=titanic,x='Pclass',y='Age',hue='Sex')
```

# 3 Box Plot(Numrical-Categorical) Comparision between age and gender 

```python
sns.boxplot(data=titanic,x='Sex',y='Age',hue='Survived')
```

# 4 Dis Plot(Numrical-Categorical) comparision between data

```python
sns.displot(data=titanic,x='Age',kde=True,color='red')
```

```python
sns.displot(data=titanic,x='Age',kind='kde',color='green')
```

```python
sns.distplot(titanic[titanic['Survived']==0]['Age'],hist=False)
sns.distplot(titanic[titanic['Survived']==1]['Age'],hist=False)
```

# OR

```python
sns.histplot(titanic[titanic['Survived']==0]['Age'],kde=True,fill=False)
sns.histplot(titanic[titanic['Survived']==1]['Age'],kde=True,fill=False)
```

```python
sns.kdeplot(titanic[titanic['Survived']==0]['Age'])
sns.kdeplot(titanic[titanic['Survived']==1]['Age'])
```

# 5 HeatMap(Categorical-Categorical)

```python
pd.crosstab(titanic['Pclass'],titanic['Survived'])
```

```python
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['Survived']))
```

```python
titanic.groupby('Sex').mean()['Survived']*100
```

# 6 Cluster Plot(Categorical-Categorical)

```python
sns.clustermap(pd.crosstab(titanic['Parch'],titanic['Survived']))
```
