```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
from sklearn import linear_model
```

```python
df=pd.read_csv('hiring2.csv')
```

```python
df.dropna()
```

```python
dummies=pd.get_dummies(df.experience)
dummies
```

```python
merged=pd.concat([df,dummies],axis='columns')
merged
```

```python
final=merged.drop(['experience'],axis='columns')
final
```
